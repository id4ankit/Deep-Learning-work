```python
# Mappings for word indices
word2idx = {word: idx for idx, word in enumerate(vocab)}
idx2word = {idx: word for word, idx in word2idx.items()}
#word2idx
```

```python
# Encoding words in text to integers
encoded_words = [word2idx[word] for word in words]
```

```python
# Prepare input and target sequences
seq_len = 5
input_seq, target_seq = [], []

for i in range(len(encoded_words) - seq_len):
    input_seq.append(encoded_words[i:i+seq_len-1])
    target_seq.append(encoded_words[i+1:i+seq_len])

# Convert sequences to tensors and move to GPU
input_seq = torch.tensor(input_seq).to(device)
target_seq = torch.tensor(target_seq).to(device)
```

```python
# Updated one_hot_encode with index checking
def one_hot_encode(sequence, vocab_size, seq_len, batch_size):
    features = np.zeros((batch_size, seq_len, vocab_size), dtype=np.float32)
    for i in range(batch_size):
        for u in range(len(sequence[i])):
            if sequence[i][u] < vocab_size:
                features[i, u, sequence[i][u]] = 1
            else:
                print(f"Warning: index {sequence[i][u]} is out of bounds for vocab_size {vocab_size}")
    return features
```

```python

class WordLevelRNN(nn.Module):
    def __init__(self, input_size, output_size, hidden_dim, n_layers=5):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers

        #RNN with five hidden layers
        self.rnn = nn.RNN(input_size, hidden_dim, n_layers, batch_first=True)

        # Fully connected layer to output the prediction
        self.fc = nn.Linear(hidden_dim, output_size)

    def init_hidden(self, batch_size):
        hidden = torch.zeros(self.n_layers, batch_size, self.hidden_dim).to(device)
        return hidden

    def forward(self, x):
        batch_size = x.shape[0]

        hidden = self.init_hidden(batch_size)
        out, hidden = self.rnn(x, hidden)

        out = self.fc(out)

        return out, hidden
```

```python
# Initialize model, loss, and optimizer
model = WordLevelRNN(input_size=len(vocab), output_size=len(vocab), hidden_dim=128, n_layers=5).to(device)
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# One-hot encode input sequence for training
input_seq = torch.tensor(one_hot_encode(input_seq.cpu().numpy(), len(vocab), seq_len-1, len(input_seq))).to(device)
```

```python
# Training loop
n_epochs = 5000
for epoch in range(1, n_epochs + 1):
    model.train()
    optimizer.zero_grad()
    output, hidden = model(input_seq)
    loss = loss_function(output.view(-1, len(vocab)), target_seq.view(-1))
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f'Epoch: {epoch}/{n_epochs}, Loss: {loss.item():.4f}')
```

```python
# Updated generate_text function
def generate_text(model, start_text, gen_length=10):
    model.eval()
    words = start_text.lower().split()
    for _ in range(gen_length):
        input_seq = [word2idx.get(word, 0) for word in words[-(seq_len - 1):]]
        input_seq = torch.tensor(one_hot_encode([input_seq], len(vocab), seq_len - 1, 1)).to(device)

        with torch.no_grad():
            output, _ = model(input_seq)
            probs = nn.functional.softmax(output[0, -1], dim=0).cpu().numpy()
            next_word_idx = np.random.choice(len(vocab), p=probs)
            words.append(idx2word[next_word_idx])

    return ' '.join(words)

```
